# Dictionaries & Pandas
- Execute the follow cell to make the st_to dictionary.

```python
#Creating dictionary from columns in dataframe
import pandas as pd
storedata = pd.read_csv('storedata_clean.csv',sep=";")

store_id = storedata['Store ID'].values.tolist()
turnover = storedata['Profit'].values.tolist()
floor = storedata['Floor Space'].values.tolist()

st_to = dict(zip(store_id, turnover))
```

## 3 Nested dictionaries
- Execute the following cell to make the dictionary st_to_fl.

```python
#Creating a nested dictionary
st_to_fl = dict(map(lambda x, y, z : (x, {"turnover":y , "floor":z}), store_id, turnover, floor))
```

- Print st_to_fl and study the structure of this 'nested' dictionary.

```python
#Get an understanding of the structure of the dictionary
st_to_fl
```

- Print the floorspace of 13132.
- Add the new store 99999 to the dictionary with a turnover of 3100000 and floor space of 15000.
- Print the floor space and turnover of this new store.
- Change the turnover of 99999 in 3200000.
- Print the turnover of the new store.

```python
def get_floorspace(stores, floorspace):
    for row in stores.values():
        if (row['floor'] == floorspace):
            return row

#print floorspace of 13132
floorspace_13132 = get_floorspace(st_to_fl, 13132)
#floorspace_13132

#add new store 99999 to the dictionary with turnover of 3100000 and floor space of 15000
store_99999 = st_to_fl
store_99999[99999] = {
    'turnover': 3100000,
    'floor': 15000
}
#store_99999

#Change turnover into 3200000
store_99999[99999].update({
    'turnover': 3200000
})
#store_99999
```
