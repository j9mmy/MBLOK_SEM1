# Exercises Seaborn
- Execute the following cell to load the dataset.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
store_date = pd.read_csv('walmart.csv', parse_dates = ['Date'])
store_date.info()
```

### General

### 1 Scatterplot temperature and turnover
- Execute the following cell.
In this cell, the dataframe st35 is being made, contain the data of store 35 from 2012. The correlation coefficient is also being calculated between the temperature and the turnover of this store.

```python
st35 = store_date[(store_date['Year'] == 2012)&(store_date['Store'] == 35)]
corr = round(np.corrcoef(st35['Temperature'],st35['Weekly_Sales'] )[0,1], 2)
corr
```

- Make a scatterplot that shows the relation between the Temperature and the Weekly_Sales, using seaborn. Make sure that dots that belong to vacation periods (IsHoliday) have a different color than those that don't.

```python
sns.scatterplot(x=st35['Weekly_Sales'], y=st35['Temperature'])
plt.show()
```

### 2 Multiple scatterplots
- Execute the following cell to make st23_27. This contains data of stores 23 up till and including 27.

```python
st23_27 = store_date[store_date['Store'].isin([23,24,25,26,27])]
st23_27.columns
```

- Make a visualization where you show a scatterplot per store, with temperature on the x-axis and turnover on the y-axis.

```python
g = sns.FacetGrid(st23_27, col='Store')
g.map(sns.scatterplot, 'Temperature', 'Weekly_Sales')
g.add_legend()
plt.show()
```

### 3 Style and colors
- Set a style, making the backgrounds of all graphs grey.-
- Set a color pallette that is suitable for people that are color-blind.

```python
g = sns.FacetGrid(st23_27, col='Store', hue='Year')
g.map(sns.scatterplot, 'Temperature', 'Weekly_Sales')
g.add_legend()

sns.set_style('dark')
sns.color_palette('colorblind')
plt.show()
```

### 4 Bar charts and box plots
- Make a bar chart, where you show the average turnover per year.
- Make a boxplot, providing insight in the turnover distribution per year.
- Make a visualization where a bar chart is shown for every storetype (Type) with the average turnover per year.
- Make a visualization where a boxplot is shown for every storetype (Type) with the turnover distribution per year.

```python
sns.barplot(st23_27, x='Year', y='Weekly_Sales')
plt.show()
```

```python
sns.boxplot(data=st23_27, x='Year', y='Weekly_Sales')
plt.show()
```

```python
g = sns.FacetGrid(st23_27, col='Type')
g.map(sns.barplot, 'Year', 'Weekly_Sales')
g.add_legend()

plt.show()
```

```python
g = sns.FacetGrid(st23_27, col='Type')
g.map(sns.boxplot, 'Year', 'Weekly_Sales')
g.add_legend()

plt.show()
```
